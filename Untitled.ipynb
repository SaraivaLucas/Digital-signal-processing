```matlab
clc;clear all;close all;
%% Parâmetros
fs=1000;
fm=2;
tempo=0:1/fs:1/fm-1/fs;
Sinal=2*sin(2*pi*fm*tempo);
l=length(Sinal);
f=zeros(5,l);
f(1,ceil(l/6):end)=Sinal(1,1:ceil(l*(1-1/6)));
f(2,ceil(l*2/6):end)=Sinal(1,1:ceil(l*(1-1*2/6)));
f(3,ceil(l*3/6)+1:end)=Sinal(1,1:ceil(l*(1-1*3/6)));
f(4,ceil(l*4/6):end)=Sinal(1,1:ceil(l*(1-1*4/6)));
f(5,ceil(l*5/6):end)=Sinal(1,1:ceil(l*(1-1*5/6)));

%% Iterações
% Iteração 1
Rxx=Sinal.*Sinal;
Y(1,1:l)=(sum(Rxx)/l);
plot(tempo,Rxx,tempo,Sinal,':',tempo,Y,'--')
legend('Rxx(\tau=0)','Sinal Original','Valor da Rxx(Tau)')
% Iteração 2
for i=1:5
figure
Rxx=Sinal.*f(i,:);
Y(1,1:l)=(sum(Rxx)/l);
plot(tempo,Rxx,tempo,Sinal,':',tempo,Y,'--')
legend(['Rxx(\tau=',num2str(i*60),')'],'Sinal Original','Valor da Rxx(Tau)')
end
```
